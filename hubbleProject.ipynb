```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from astropy.io import fits

file = '/home/ryan/Documents/cosmology/Hubble-Constant/data/Galaxy+spectra+and+distances+-+Sheet1+(1).csv' #file obtained online
data = pd.read_csv(file)
c = 299792 #speed of light
```

# Equations for Redshift, velocity, and gaussian calculations

```python
def redshift(lambda_o, lambda_e):
    return (lambda_o - lambda_e) / lambda_e

def v_r(z):
    v = z * c
    return v

def gauss(x, A, mu, sig, c):
    return A * np.exp(-(x - mu)**2 / (2 * sig**2)) + c
```

# Velocity vs Distance Plot and Hubble Constant Value

```python
x = data['Distance (Mpc)'].to_numpy()[:25]
y = data['Velocity v (km/s)'].to_numpy()[:25]
a, b = np.polyfit(x, y, 1)

plt.scatter(x,y)
plt.plot(x, a*x+b)
plt.xlabel("Distance (Mpc)")
plt.ylabel("Velocity v (km/s)")

print("Slope = ", a, "km/s/Mpc")
```

# Age of the Universe

```python
H0 = 73.190390
H0_sec = H0 * 1 / (3.086e19)
H0_yr = H0_sec * (3.1536e7)

age = (1 / H0_yr) /1e9
print(f'Estimated age of universe: {age:.2f} billion years')
```

# NGC 99

```python
hdul = fits.open('/home/ryan/Documents/cosmology/Hubble-Constant/data/fits/NGC99.fits')
spec = hdul[1].data['flux']
waves = 10**hdul[1].data['loglam']

fig = plt.figure()
fig.set_size_inches(25,10)

plt.plot(waves, spec)
plt.xlabel("wavelength (angstroms)")
plt.ylabel("flux")
plt.title("Entire spectrum")
```

```python
x_min = 6600
x_max = 6800

mask = (waves >= x_min) & (waves <= x_max)
waves2 = waves[mask]
spec2 = spec[mask]

fig = plt.figure()
fig.set_size_inches(25,10)
plt.plot(waves2, spec2)
plt.ylabel("flux")
plt.title("Focused on H alpha")
```

```python
guess = [130, 6680, 1.5, 130]
fit_params, _ = curve_fit(gauss, waves2, spec2, guess)
print(fit_params[0])
print(fit_params[3])
```

```python
plt.plot(waves2, spec2, label = "spectral data")
plt.plot(waves2, gauss(waves2, *fit_params), color='red', label = "fit gaussian")
plt.legend()
plt.xlabel("wavelength (angstroms)")
plt.ylabel("flux")
plt.title("Gauss Fit")
```

```python
lambda_o = fit_params[1]
lambda_r = 6563
z = redshift(lambda_o, lambda_r)

v = v_r(z)

print(z)
print(v)
```

```python
guess = [140, 6690, 1.5, 70]
fit_params, _ = curve_fit(gauss, waves2, spec2, guess)
print(fit_params[0])
print(fit_params[3])

plt.plot(waves2, spec2, label = "spectral data")
plt.plot(waves2, gauss(waves2, *fit_params), color='red', label = "fit gaussian")
plt.legend()
plt.xlabel("wavelength (angstroms)")
plt.ylabel("flux")
plt.title("Gauss Fit")
```

```python
lambda_o = fit_params[1]
lambda_r = 6563
z = redshift(lambda_o, lambda_r)

v = v_r(z)

print(z)
print(v)
```
